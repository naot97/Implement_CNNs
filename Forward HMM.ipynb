#Excercise of Fordward Hidden Markov Model

```
import numpy as np
```

##Data

```
n_x = 4
n_y = 4
pyy = np.zeros((n_y,n_y))
pyx = np.zeros((n_y,n_x))

#P(y->y) 
#y: 0-Party 1-Tv 2-Pub 3-Study
pyy[0] = np.array([0.05, 0.7, 0.05, 0.2])
pyy[1] = np.array([0.1,  0.4, 0.3, 0.2 ])
pyy[2] = np.array([0.1,  0.6, 0.05, 0.25])
pyy[3] = np.array([0.25, 0.3, 0.4, 0.05])

#p(x|y) 
#y : 0-Party 1-Tv 2-Pub 3-Study; 
#x : 0-tired 1-hungover 2-scared 3-fine;
pyx[0] = np.array([0.3, 0.4, 0.2, 0.1])
pyx[1] = np.array([0.2, 0.1, 0.2, 0.5])
pyx[2] = np.array([0.4, 0.2, 0.1, 0.3])
pyx[3] = np.array([0.3, 0.05, 0.3, 0.35])
```

##Implement

```
def mapCharacter(n):
	return {
		0: 'Party',1: 'Tv',2: 'Pub',3: 'Study'
				}.get(n,'null')
```

```
input = np.array([0,0,2])
```

```
o = np.ones((n_y,)) / n_y
temp = np.zeros((n_y,))
o = o * pyx[:, input[0]]
print('output of forward model hmm :')
print('layer 0', o)
for i in range(1, len(input)):
  x = input[i]
  for y in range(n_y):
    temp[y] = np.dot(pyy[:,y],o) * pyx[y,x]
  o = temp
  print('layer %d' % (i), o)
```
